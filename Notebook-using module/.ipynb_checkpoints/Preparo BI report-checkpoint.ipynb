### Script final 

```python
import pandas as pd


filename = "C:/Users/a844026/Downloads/Planning des sessions Intra Pré Programmées 2023 - mise à jour 16 01 23.xlsx"
df = pd.read_excel(filename, sheet_name = 0, header=4, dtype='object') # read all sheets 


def renaming_cols(df):
    mois = []
    for col in df.columns:
        if 'Dates formation' in col:
            new_col = ' '.join(col.split(' ')[-2:])
            df.rename(columns = {col: new_col}, inplace = True)
            mois.append(new_col)
        elif col == "Unnamed: 0":
            df.rename(columns = {col: "Formation"}, inplace = True)
        elif col == "Unnamed: 25":
            df.rename(columns = {col: "ID"}, inplace = True)
        elif col == "Unnamed: 26":
            df.rename(columns = {col: "urls"}, inplace = True)
        # else:
            # df[col] = pd.to_datetime(df[col])
            
    new_col = {"Dates limite d'inscription ": 'janv deadline', "Dates limite d'inscription .1": 'fev deadline',
               "Dates limite d'inscription .2": 'mars deadline', "Dates limite d'inscription .3": 'avr deadline',
               "Dates limite d'inscription .4": 'mai deadline', "Dates limite d'inscription .5": 'juin deadline',
               "Dates limite d'inscription .6": 'juil deadline', "Dates limite d'inscription .7": 'aout deadline',
               "Dates limite d'inscription .8": 'sept deadline', "Dates limite d'inscription .9": 'oct deadline',
               "Dates limite d'inscription .10": 'nov deadline', "Dates limite d'inscription .11": 'dec deadline'}

    df.rename(columns = new_col, inplace = True)
    return df, mois

df, mois = renaming_cols(df)

# reshape DataFrame from wide format to long format
id_vars=['ID', 'Formation', 'urls'] 
df2 = pd.melt(df, id_vars=id_vars, value_vars=mois,
             var_name='MOIS', value_name='DATE')
df2 = df2.sort_values(by =['ID', 'Formation', 'urls'] )
df2.reset_index(drop=True, inplace=True)
 
temp = pd.melt(df, id_vars=id_vars, value_vars=new_col.values(),
             var_name='MOIS', value_name='DEADLINE')
temp = temp.sort_values(by =['ID', 'Formation', 'urls'] )
temp.reset_index(drop=True, inplace=True)

df2['DEADLINE'] = temp.DEADLINE

def get_place(row):
    if str(row) != 'nan':
        val = row.split()
        if len(val) == 4:
            return val[-1]
        else:
            return val[-2]
    else:
        return row
    
def get_date(row):
    if str(row) != 'nan':
        val = row.split()
        if len(val) == 4:
            return ' '.join(row.split()[:-1])
        else:
            return ' '.join(row.split()[:-2])
    else:
        return row

df2['PLACES'] = df2.apply(lambda x: get_place(x.DATE), axis=1)
df2['DATE'] = df2.apply(lambda x: get_date(x.DATE), axis=1)
```
