```python
slices['time'] = a['Slices']
slices['time'] = slices['time'].apply(pd.to_datetime, format='%Y-%m-%dT%H:%M:%S') 
slices['Slice_duration'] = slices['Slice_duration'].apply(pd.Timedelta)

slices['time_end'] = slices['time'] + slices['Slice_duration']
# slices['time_end'] = slices['time'].shift(periods=-1, fill_value=np.nan)
# slices['time_end'][-1:] = slices['Slice_duration'][-1:] + slices['time'][-1:] 
slices["alarms"] = "NAN"
print(slices.dtypes)

slices
```

```python
n_slice = 1
# slice start and end
start_date = slices.time[n_slice]
end_date  = slices.time_end[n_slice]
result = data_algo[(data_algo.TSTART >= start_date) & (data_algo.TSTART <= end_date)]
result
```

```python
result.Alarm_number.values
```

```python
def get_alarms(sta,end):
    return data_algo[(data_algo.TSTART >= sta) & (data_algo.TSTART <= end)].Alarm_number.values

def get_nb_alarms(sta,end):
    return len(data_algo[(data_algo.TSTART >= sta) & (data_algo.TSTART <= end)].Alarm_number.values)

slices['alarms'] = slices.apply(lambda x: get_alarms(x.time, x.time_end), axis=1)
slices['nb_alarms'] = slices.apply(lambda x: get_nb_alarms(x.time, x.time_end), axis=1)

slices
```

```python
data['Slice_duration'] = data['Slice_duration'].apply(pd.to_timedelta) 
data
```

```python
## slice_start

u = s1_start
temp1 = []
temp1.append(u)
#print(f's{1}_start ={u}')
for i in range(1,len(data)): 
    r = data["Slice_duration"][i]
    u = u + r
    temp1.append(u)

data['s_start'] = temp1
```

```python
data['s_end'] = data['s_start'].shift(periods=-1, fill_value=data_algo["TEND"].max())


def _get_alarms(sta, end):
    r = data_algo[(data_algo.TSTART >= sta) &
                  (data_algo.TSTART < end) | (data_algo.TEND <= end) ]
    return r.Alarm_number.values

def _get_nb_alarms(sta,end):
    return len(data_algo[(data_algo.TSTART >= sta) & (data_algo.TSTART < end)].Alarm_number.values)

def get_alarms(sta, end):
    r = data_algo[(data_algo.TSTART < end) & (data_algo.TEND > end)]
    return r.Alarm_number.values

def get_nb_alarms(sta,end):
    return len(data_algo[(data_algo.TSTART <= end) & (data_algo.TEND > end)].Alarm_number.values)


data['alarms'] = data.apply(lambda x: get_alarms(x.s_start, x.s_end), axis=1)
data['nb_alarms'] = data.apply(lambda x: get_nb_alarms(x.s_start, x.s_end), axis=1)

data
```
